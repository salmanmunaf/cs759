```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../cuda/results/length.csv")
df_serial = pd.read_csv("../serial/results/length.csv")
plt.plot(df["Input"], df["Time"], label='CUDA')
plt.plot(df_serial["Input"], df_serial["Time"], label='Serial')
# plt.xticks(np.arange(1,27))
# plt.xlim(1,26)
plt.xlabel("Password length")
plt.ylabel("Time (ms)")
plt.legend()
# plt.plot(kind='scatter', x='Input', y='Time', color='red', xticks=np.arange(5,25), 
#         xlabel='Input Size: 2^n', ylabel='Time (ms)')
# plt.show()
plt.savefig('length.pdf')
```

```python
plt.plot(df["Input"], np.log2(df["Time"]), label='CUDA')
plt.plot(df_serial["Input"], np.log2(df_serial["Time"]), label='Serial')
# plt.xticks(np.arange(1,27))
# plt.xlim(1,26)
plt.xlabel("Password length")
plt.ylabel("Time (ms)")
plt.legend()
# plt.plot(kind='scatter', x='Input', y='Time', color='red', xticks=np.arange(5,25), 
#         xlabel='Input Size: 2^n', ylabel='Time (ms)')
# plt.show()
plt.savefig('length_log.pdf')
```

```python
df = pd.read_csv("../cuda/results/threads_per_block.csv")
df.plot(kind='scatter', x='Threads', y='Time', color='red', xticks=np.arange(1,10), 
        xlabel='Number of Threads per block: 2^n', ylabel='Time (ms)')
# plt.show()
plt.savefig('threads_per_block.pdf')
```

```python
df = pd.read_csv("../cuda/results/devices.csv")
df.plot(kind='scatter', x='Devices', y='Time', color='red', xticks=np.arange(1,6), 
        xlabel='Number of GPU devices', ylabel='Time (ms)')
# plt.show()
plt.savefig('devices.pdf')
```
